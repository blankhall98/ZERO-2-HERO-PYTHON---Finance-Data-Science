# Crecimiento

```python
# PAQUETERIAS
import numpy as np
import matplotlib.pyplot as plt
```

```python
# BASE DE DATOS
year = [1800,1850,1900,1950,2000]
pop = [984,1280,1650,2540,6140]
```

```python
# Valores estadísticos
def cambio_absoluto(values):
    token = []
    for i in range(len(pop)-1):
        ca = values[i+1] - values[i]
        token.append(ca)
    return token
```

```python
ca = cambio_absoluto(pop)
year_adj = year[1:]
```

```python
cambio_promedio = np.mean(ca)
print(f'Cambio Promedio: {cambio_promedio}')

unidad_tiempo = 50
cambio_promedio_anual = cambio_promedio/unidad_tiempo
print(f'Cambio Promedio: {cambio_promedio_anual}')
```

```python
year_adj
```

```python
tendencia_promedio = []
for año in year:
    y = cambio_promedio_anual*(año - year[0]) + pop[0]
    tendencia_promedio.append(y)
print(tendencia_promedio)
```

```python
# Gráfica 1
plt.figure(figsize = (10,5))

plt.plot(year,pop,'bo--',label='Datos Reales')
plt.plot(year,tendencia_promedio,label='Tendencia Lineal')
plt.title('Población Mundial Total 1800-2000')
plt.xlabel('Año')
plt.ylabel('Población Total')

plt.legend()

plt.show()
```

```python
plt.figure()

plt.title('Cambio Absoluto en la Población')
plt.plot(year_adj,ca,'o--')

plt.show()
```
